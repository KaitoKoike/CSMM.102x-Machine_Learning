```python
import numpy as np
import sys
import pandas as pd
```

```python
n = 10

num_inputs = 3

A = 5
B = 8
C = 2
```

```python
x = np.empty([0,num_inputs-1])

for a in range(n):
    for i in range(n):
        x = np.append(x, [[a, i]], axis=0)
    
x = np.hstack((x, np.ones((n**(num_inputs-1),1))))
```

```python
def f(x):
    return(A*x[0] + B*x[1])

y = np.array([])

for i in x:
    y = np.append(y, f(i))
    
y = y.reshape(-1, 1)
```

```python
print x, y
```

```python
lambda_input = int(3)
sigma2_input = float(16)

X_train = x
y_train = y
```

```python
X_test = np.random.randint(20, size=(n*2, num_inputs-1))
X_test = np.hstack((X_test, np.ones((n*2,1))))
print X_test
```

```python
def part1(X, y, lam):
    
    lambda_i = lam*np.identity(num_inputs)
    
    XX = np.matmul(np.transpose(X), X)
    
    XX_recip = np.linalg.matrix_power(lambda_i + XX, -1)
    
    Xy = np.matmul(np.transpose(x), y.reshape(-1,1))
    
    wRR_prediction = np.matmul(XX_recip, Xy)
    
    return wRR_prediction

wRR = part1(X_train, y_train, lambda_input)
np.savetxt("wRR_" + str(lambda_input) + ".csv", wRR, delimiter="\n")

print wRR
print A, B
```

```python
def part2(X, x0):
    
    sigma = sigma2_input**(0.5)
    
    x0 = np.hstack((x0, np.arange(x0[:,0].size).reshape(-1,1)))
    
    def get_index(X, x0):
        
        sigma2_list = np.array([])

        lam_i = lambda_input*np.eye(num_inputs)
        sigm2x = sigma**(-2)*np.transpose(X)
        Sig = np.linalg.matrix_power(lam_i + np.matmul(sigm2x, X),-1)
        
        for i in range(x0[:,0].size):
                        
            sigma20 = sigma2_input + np.matmul(np.matmul(np.transpose(x0[i, 0:-1]), Sig), x0[i, 0:-1])

            sigma2_list = np.append(sigma2_list, sigma20)
    
        return(np.argmax(sigma2_list))
        
    X_known = np.copy(X)
    X_unkown = np.copy(x0)
    
    indexes = np.empty([0, 1])
    
    for a in range(10):
                
        index = get_index(X_known, X_unkown)
                
        X_known = np.append(X_known, X_unkown[index, 0:-1].reshape(1,num_inputs), axis=0)
        
        indexes = np.append(indexes, X_unkown[index, -1])
        
        X_unkown = np.delete(X_unkown, index, 0)
        
    return indexes

active = part2(X_train, X_test)+1

print active
np.savetxt("active_" + str(lambda_input) + "_" + str(int(sigma2_input)) + ".csv", active, delimiter=",") # write output to file
```
